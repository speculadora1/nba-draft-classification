```python
# import dependencies
from sportsreference.ncaab.roster import Player
import pandas as pd
```

```python
# load player id map
ids = pd.read_csv('../data/id_map.csv')
ids.head()
```

```python
ids.dropna(subset = ['cbb_id'], inplace = True)
ids_clean = ids[ids.pro_id != 'fernaru01']
```

```python
ncaa_data = pd.DataFrame()

for p_id in ids_clean['cbb_id']:
    
    try:
        print(f'Retrieving data for {p_id}')
        
        # get player object for id
        plyr = Player(p_id)

        # get player dataframe
        player_df = plyr.dataframe

        # reset index
        player_df = player_df.reset_index()

        # append to dataframe
        ncaa_data = ncaa_data.append(player_df, ignore_index = True)
        
    except:
        print(f'No data found for {p_id}')
        
# remove rows for career totals, rename former index column, move player_id column to first position
ncaa_data = ncaa_data[ncaa_data.level_0 != 'Career']
ncaa_data = ncaa_data.rename(columns = {'level_0': 'season'})
ncaa_data = ncaa_data[['player_id'] + [col for col in ncaa_data.columns if col != 'player_id']]
```

```python
ncaa_data.to_csv('../data/ncaa_stats.csv', index = False)
```
