```python
from sportsreference.ncaab.roster import Player
from sportsreference.ncaab.roster import Roster
import pandas as pd
```

```python
ids = pd.read_csv('../data/id_map.csv')
ids.head()
```

```python
ids_dropped = ids.dropna(subset=['cbb_id'])
ids_dropped = ids_dropped[ids_dropped.pro_id != 'fernaru01']
ids_dropped
```

```python
assist_percentage = []
assists = []
box_plus_minus = []
block_percentage = []
blocks = []
conference = []
defensive_rebound_percentage = []
defensive_rebounds = []
effective_field_goal_percentage = []
field_goal_attempts = []
field_goals = []
free_throw_attempt_rate = []
free_throw_attempts = []
free_throw_percentage = []
free_throws = []
minutes_played = []
name = []
offensive_rebound_percentage = []
offensive_rebounds = []
personal_fouls = []
player_id = []
points = []
steal_percentage = []
steals = []
three_point_attempt_rate = []
three_point_attempts = []
three_point_percentage = []
three_pointers = []
total_rebound_percentage = []
total_rebounds = []
true_shooting_percentage = []
turnover_percentage = []
turnovers = []
two_point_attempts = []
two_point_percentage = []
two_pointers = []
usage_percentage = []
win_shares = []
```

```python
for index, row in ids_dropped.iterrows():
    final_season = f"{row['season']-1}-{str(row['season'])[2:4]}"
    player_reference = Player(row['cbb_id'])
    new_player = player_reference(final_season)
    assist_percentage.append(new_player.assist_percentage)
    assists.append(new_player.assists)
    box_plus_minus.append(new_player.box_plus_minus)
    block_percentage.append(new_player.block_percentage)
    blocks.append(new_player.blocks)
    conference.append(new_player.conference)
    defensive_rebound_percentage.append(new_player.defensive_rebound_percentage)
    defensive_rebounds.append(new_player.defensive_rebounds)
    effective_field_goal_percentage.append(new_player.effective_field_goal_percentage)
    field_goal_attempts.append(new_player.field_goal_attempts)
    field_goals.append(new_player.field_goals)
    free_throw_attempt_rate.append(new_player.free_throw_attempt_rate)
    free_throw_attempts.append(new_player.free_throw_attempts)
    free_throw_percentage.append(new_player.free_throw_percentage)
    free_throws.append(new_player.free_throws)
    minutes_played.append(new_player.minutes_played)
    name.append(new_player.name)
    offensive_rebound_percentage.append(new_player.offensive_rebound_percentage)
    offensive_rebounds.append(new_player.offensive_rebounds)
    personal_fouls.append(new_player.personal_fouls)
    player_id.append(new_player.player_id)
    points.append(new_player.points)
    steal_percentage.append(new_player.steal_percentage)
    steals.append(new_player.steals)
    three_point_attempt_rate.append(new_player.three_point_attempt_rate)
    three_point_attempts.append(new_player.three_point_attempts)
    three_point_percentage.append(new_player.three_point_percentage)
    three_pointers.append(new_player.three_pointers)
    total_rebound_percentage.append(new_player.total_rebound_percentage)
    total_rebounds.append(new_player.total_rebounds)
    true_shooting_percentage.append(new_player.true_shooting_percentage)
    turnover_percentage.append(new_player.turnover_percentage)
    turnovers.append(new_player.turnovers)
    two_point_attempts.append(new_player.two_point_attempts)
    two_point_percentage.append(new_player.two_point_percentage)
    two_pointers.append(new_player.two_pointers)
    usage_percentage.append(new_player.usage_percentage)
    win_shares.append(new_player.win_shares)
    print(f"{row['cbb_id']}")
```

```python
final_df = pd.DataFrame(list(zip(assist_percentage, assists, box_plus_minus, block_percentage, blocks, conference, defensive_rebound_percentage, defensive_rebounds, effective_field_goal_percentage, field_goal_attempts, field_goals, free_throw_attempt_rate, free_throw_attempts, free_throw_percentage, free_throws, minutes_played, name, offensive_rebound_percentage, offensive_rebounds, personal_fouls, player_id, points, steal_percentage, steals, three_point_attempt_rate, three_point_attempts, three_point_percentage, three_pointers, total_rebound_percentage, total_rebounds, true_shooting_percentage, turnover_percentage, turnovers, two_point_attempts, two_point_percentage, two_pointers, usage_percentage, win_shares)), columns=['assist_percentage', 'assists', 'box_plus_minus', 'block_percentage', 'blocks', 'conference', 'defensive_rebound_percentage', 'defensive_rebounds', 'effective_field_goal_percentage', 'field_goal_attempts', 'field_goals', 'free_throw_attempt_rate', 'free_throw_attempt', 'free_throw_percentage', 'free_throws', 'minutes_played', 'name', 'offensive_rebound_percentage', 'offensive_rebounds', 'personal_fouls', 'player_id', 'points', 'steal_percentage', 'steals', 'three_point_attempt_rate', 'three_point_attempts', 'three_point_percentage', 'three_pointers', 'total_rebound_percentage', 'total_rebounds', 'true_shooting_percentage', 'turnover_percentage', 'turnovers', 'two_point_attempts', 'two_point_percentage', 'two_pointers', 'usage_percentage', 'win_shares'])
```

```python
final_df.to_csv('../data/ncaa_stats.csv')
```
